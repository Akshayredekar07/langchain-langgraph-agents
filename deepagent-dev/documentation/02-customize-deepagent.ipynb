```python
import os
from typing import Literal
from dotenv import load_dotenv
from langchain.chat_models import init_chat_model
from langchain_core.tools import tool
from tavily import TavilyClient
from deepagents import create_deep_agent

load_dotenv()

# ── API Keys ──────────────────────────────────────────────────────────────────
def _require_env(name: str) -> str:
    value = os.getenv(name)
    if value is None:
        raise ValueError(f"Missing required environment variable: {name}")
    return value

os.environ["HUGGINGFACEHUB_API_TOKEN"] = _require_env("HUGGINGFACEHUB_API_TOKEN")
os.environ["TAVILY_API_KEY"]           = _require_env("TAVILY_API_KEY")

# ── Option 1: Shorthand (simple model string) ────────────────────────────────
# agent = create_deep_agent(
#     model="gpt-4",
# )

# ── Option 2: Explicit init_chat_model (recommended for fine-grained control) ─
model = init_chat_model(
    # model="microsoft/Phi-3-mini-4k-instruct",
    model="claude-sonnet-4-6",
    model_provider="anthropic",
    # model_provider="huggingface",
    temperature=0.7,
    max_tokens=1024,
)

# ── Custom Tool ───────────────────────────────────────────────────────────────
tavily_client = TavilyClient(api_key=os.environ["TAVILY_API_KEY"])

@tool
def web_search(
    query: str,
    max_results: int = 5,
    topic: Literal["general", "news", "finance"] = "general",
) -> dict:
    """Search the web for up-to-date information."""
    return tavily_client.search(
        query,
        max_results=max_results,
        topic=topic,
        include_raw_content=False,
    )

# ── Build Agent (with all knobs) ──────────────────────────────────────────────
agent = create_deep_agent(
    model=model,                          # BaseChatModel or "provider:model" string
    tools=[web_search],                   # your custom tools (built-ins auto-included)
    system_prompt="You are a research assistant. Always plan before acting.",
    debug=True,                           # prints tool calls & state transitions
)

# ── Invoke (single-shot) ──────────────────────────────────────────────────────
result = agent.invoke({
    "messages": [
        {"role": "user", "content": "Research the latest trends in Vectorless RAG systems and write a summary."}
    ]
})

print(result["messages"][-1].content)
```

```python
# ── Streaming (recommended for long tasks) ───────────────────────────────────
for chunk in agent.stream({
    "messages": [
        {"role": "user", "content": "What are the top open-source LLMs in 2025?"}
    ]
}):
    # chunk keys: 'agent', 'tools', 'filesystem', etc.
    for key, val in chunk.items():
        if "messages" in val:
            for msg in val["messages"]:
                print(f"[{key}] {msg.content}", end="", flush=True)
```
